```python
from selenium import webdriver
import time
import pandas as pd
from pathlib import Path
```

```python
cwd = Path.cwd()
cwd
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
browser = webdriver.Chrome(chrome_options=chrome_options)
```

```python
song_urls = []
browser.get('https://music.163.com/#/discover/toplist?id=3778678')
browser.switch_to.frame('g_iframe')
items = browser.find_elements_by_xpath('//table[@class = "m-table m-table-rank"]/tbody/tr//span[@class = "txt"]/a')
for item in items:
    href = item.get_attribute('href')
    song_urls.append(href)
```

```python
def func_comments(i):
    comments = []
    likes = []
    time = []
    browser.get(song_urls[i])
#     time.sleep(1)
    try:
        browser.switch_to.frame('g_iframe')  #记得切换到所要找的元素所在的frame，不然会找不到元素，这里有的网页不需要切换，所以使用try
    except:
        pass
    song_name = browser.find_element_by_xpath('//div[@class = "tit"]/em')
    comments_num = browser.find_element_by_xpath('/html/body/div[3]/div[1]/div/div/div[2]/div/div[1]/span/span').text
    comments_cont = browser.find_elements_by_xpath('//div[@class = "cmmts j-flag"]//div[@class = "cntwrap"]/div/div[@class = "cnt f-brk"]')
    likes_num = browser.find_elements_by_xpath('//div[@class = "cmmts j-flag"]//div[@class = "cntwrap"]/div[@class="rp"]/a[1]')
    comments_time = browser.find_elements_by_xpath('//div[@class = "cmmts j-flag"]//div[@class = "cntwrap"]/div[@class="rp"]/div[@class="time s-fc4"]')

    for comment_cont in comments_cont:
        comments.append(comment_cont.text + '\n')
    comments_df = pd.DataFrame(comments,columns=['comments'])
    comments_15 = comments_df['comments'].loc[0:14]
    comments_split = comments_15.str.split('：',expand=True)[1]
    comments_split = pd.DataFrame(comments_split)
    
    for like_num in likes_num:
        likes.append(like_num.text)
    likes_df = pd.DataFrame(likes,columns=['likes'])
    likes_15 = likes_df['likes'].loc[0:14]
    likes_split = likes_15.str.split('(',expand=True)[1].str.split(')',expand=True)[0]
    likes_split = pd.DataFrame(likes_split)
    
    for comment_time in comments_time:
        time.append(comment_time.text)
    time_df = pd.DataFrame(time,columns=['time'])
    time_15 = time_df['time'].loc[0:14]
    
    comments_print = pd.concat([comments_split,likes_split,time_15],axis=1)
    comments_print['comments_num'] = comments_num
    print('当前正在抓取歌曲“{}”的评论'.format(song_name.text))
    comments_print.to_csv(song_name.text + '.csv',encoding='utf_8_sig')    
```

```python
for i in range(len(song_urls)):
    func_comments(i)
```
